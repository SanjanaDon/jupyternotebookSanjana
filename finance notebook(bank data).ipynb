```python
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_pickle('all_banks') 
```

```python
df
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
```

```python
BAC = data.DataReader('BAC','yahoo',start,end)
C = data.DataReader('C','yahoo',start,end)
GS = data.DataReader('GS','yahoo',start,end)
JPM = data.DataReader('JPM','yahoo',start,end)
MS = data.DataReader('MS','yahoo',start,end)
WFC= data.DataReader('WFC','yahoo',start,end)
```

```python
BAC
```

```python
tickers = 'BAC C GS JPM MS WFC'.split()
tickers
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
#max close price for each bank during this time period(multi level index method)
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
#max close price for each bank during this time period(.xs method)
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
#the creation of an empty dataframe called 'returns', fill with each banks returns on stocks
returns=pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return']=bank_stocks[tick]['Close'].pct_change() #pct_change()calculates percentage change
```
